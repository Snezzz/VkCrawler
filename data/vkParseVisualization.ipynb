```python
import numpy as np
import matplotlib as mp
from matplotlib import pyplot as plt
import csv
```

```python
def create_graphic(filename):
    values = []
    with open(filename + '.csv', newline='') as File:  
        reader = csv.reader(File)

        for row in reader:
            if row[0] == "day":
                continue
            values.append([str(row[0]),int(row[1])])
    x = []
    y = []
    
    values = sorted(values,key=lambda y_p: y_p[0])
   
    for day in range(0,len(values)):
        x.append(values[day][0])
        y.append(values[day][1])
    
    fig, ax = plt.subplots()
    ax.bar(x, y)

    ax.set_facecolor('seashell')
    fig.set_facecolor('floralwhite')
    fig.set_figwidth(12)    #  ширина Figure
    fig.set_figheight(6)    #  высота Figure

    plt.show()
    fig.savefig(filename + '.jpg')
    
```

```python
create_graphic('mguData')
```

```python
create_graphic('spbuData')
```
